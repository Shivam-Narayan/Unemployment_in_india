# Data Analysis Part

```python
plt.figure(figsize=(20,10)) #Unemployment rate in india
sns.lineplot(data=data_set,x='Date',y=" Estimated Unemployment Rate (%)",color='blue')
plt.title("Unemployment rate in India",fontsize="20")
```

```python
sns.barplot(data=data_set,x=" Estimated Unemployment Rate (%)",y='Region',color='cyan')
```

```python
sns.barplot(data=data_set,x='Region.1',y=" Estimated Unemployment Rate (%)",color='yellow')
```

```python
plt.figure(figsize=(20,10))
data_set.columns=['Region','Date','Frequency','Estimated Unemployment Rate (%)','Estimated Employment', 'Estimated Employed','Region.1 ','longitude','latitude']
plt.title("Unemployment in india rural and urban area.",fontsize='10')
sns.histplot(x='Estimated Unemployment Rate (%)',hue='Region',data=data_set)
plt.show()
```

```python
sns.heatmap(data_set.corr(),annot=True)
```
